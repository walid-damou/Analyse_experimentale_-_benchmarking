```python
import matplotlib.pyplot as plt

def read_and_plot_instances(file_path):
    with open(file_path, 'r') as file:
        instance_lines = []
        instance_number = 1
        
        for line in file:
            if line.strip() == '%-':
                if instance_lines:  # S'il y a des donnes accumules pour une instance
                    # Convertir les lignes avec des valeurs numriques spares par des tabulations
                    data = [list(map(float, value.split('\t'))) for value in instance_lines if value.strip()]
                    # Transposer la liste pour sparer les colonnes
                    columns = list(zip(*data))
                    
                    # Cration du graphique pour chaque colonne de donnes
                    plt.figure(figsize=(10, 5))
                    for i, col in enumerate(columns):
                        plt.plot(col, marker='o', linestyle='-', label=f'Colonne {i+1}')
                    
                    plt.title(f"Visualisation de l'instance {instance_number}")
                    plt.xlabel("Index")
                    plt.ylabel("Valeur")
                    plt.legend()
                    plt.grid(True)
                    plt.show()
                    
                    instance_number += 1
                    instance_lines = []  # Rinitialiser pour la nouvelle instance
            else:
                instance_lines.append(line.strip())  # Accumuler les valeurs de l'instance en cours

file_path = 'bbob-biobj_f01_d05_hyp.dat'
read_and_plot_instances(file_path)
```
